```python
from PIL import Image
import numpy as np
import gradio as gr
import tensorflow as tf
import os
```

```python
model = tf.keras.models.load_model(os.path.join("../trained model/best_model_version.keras")) 
```

```python
model.input_shape
```

```python
def predict_probs(image):
    image = image.resize((180, 180))  
    image_array = tf.keras.utils.img_to_array(image)
    image_array = np.expand_dims(image_array, axis=0)

    prediction = model.predict(image_array)
    return prediction[0][np.argmax(prediction)]

interface = gr.Interface(
    fn=predict_probs,
    inputs=gr.Image(type="pil"),
    outputs=gr.Label(),
    title="Maize classification disease",
    description="Upload an image and see predictions."
)
interface.launch()
```
